## Exercise 2: Cleaning invalid Values

The dataset contains columns with inconsistent and incorrect values that could affect the accuracy of our analysis. Therefore, we need to clean these columns to ensure that they only contain valid data.

Note that this exercise will focus only on cleaning inconsistent values and will not involve handling null values (NaN or None).

*Hint*:
- *Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".*
- *State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"*
- *In education, "Bachelors" could be replaced by "Bachelor"*
- *In Customer Lifetime Value, delete the `%` character*
- *In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"*

```python
print(data['gender'].unique())
print()

print(data['state'].unique())
print()

print(data['education'].unique())
print()

print(data['customer_lifetime_value'].head())
print()

print(data['vehicle_class'].unique())
```

```python
# Your code here
# Gender column contains various inconsistent 
# values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".

data['gender'] = data['gender'].replace({'Femal': 'F', 'female': 'F', 'Male': 'M'}).str.upper()
data
```

```python
print(data['gender'].unique())
```

```python
# State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"

change_state_name = {'AZ': 'Arizona', 'Cali': 'California', 'WA': 'Washington'}
data['state'] = data['state'].replace(change_state_name)
data
```

```python
print(data['state'].unique())
print()
```

```python
# In education, "Bachelors" could be replaced by "Bachelor"

data['education'] = data['education'].replace({'Bachelors': 'Bachelor'})
data
```

```python
print(data['education'].unique())
print()
```

```python
# In Customer Lifetime Value, delete the % character

data['customer_lifetime_value'] = data['customer_lifetime_value'].str.replace('%', '').astype(float) / 100
data
```

```python
# In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"

vehicle_name_change = {'Sports Car': 'Luxury', 'Luxury SUV': 'Luxury', 'Luxury Car': 'Luxury'}
data['vehicle_class'] = data['vehicle_class'].replace(vehicle_name_change)
data
```

```python
print(data['vehicle_class'].unique())
print()
```

```python
# Your code here
numerical_columns = data.select_dtypes(include=['int64', 'float64']).columns

for column in numerical_columns:
    data[column] = data[column].fillna(data[column].median())
```

```python
data
```

```python
categorical_columns = data.select_dtypes(include=['object']).columns
```

```python
for column in categorical_columns:
    mode_value = data[column].mode()[0]
    data[column] = data[column].fillna(mode_value)
```

```python
contingency_table = pd.crosstab(data['education'], data['vehicle_class'])

contingency_table.plot(kind='bar', stacked=True, figsize=(15, 9))
plt.title('Vehicle Class Distribution Across Education Levels')
plt.ylabel('Count')
plt.xlabel('Education')
plt.xticks(rotation=45)
plt.legend(title='Vehicle Class')
plt.tight_layout()
plt.show()
```
