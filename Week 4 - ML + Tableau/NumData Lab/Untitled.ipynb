```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('categorical.csv')
df
```

```python
df = df.drop('customer', axis=1)
```

```python
for column in df.select_dtypes(include='object').columns:
    plt.figure(figsize=(10,5))
    sns.countplot(data=df, x=column)
    plt.title(f'Distribution of {column}')
    plt.xticks(rotation=75)
    plt.show()
```

```python
for column in df.columns:
    display(f'{column}:')
    display(df[column].value_counts())
    print('\n')
```

```python
plt.figure(figsize=(15, 7))
df['effective_to_date'].value_counts().sort_index().plot()
plt.title('Distribution Over Time')
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
df['effective_to_date'] = pd.to_datetime(df['effective_to_date'])

# Extract day, week, and month
df['day'] = df['effective_to_date'].dt.day
df['week'] = df['effective_to_date'].dt.isocalendar().week
df['month'] = df['effective_to_date'].dt.month
```

```python
plt.figure(figsize=(15, 7))
df['effective_to_date'].value_counts().sort_index().plot()
plt.title('Distribution Over Time')
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```
